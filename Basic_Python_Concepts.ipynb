## 2. If-Else and Elif Statements

```python
x = 5
if x > 0:
    print('Positive number')
else:
    print('Non-positive number')
```

```python
x = int(input('Enter a number: '))
if x > 0:
    print('Positive')
elif x < 0:
    print('Negative')
else:
    print('Zero')
```

```python
marks = 75
if marks >= 80:
    print('Grade A')
elif marks >= 50:
    print('Grade B')
else:
    print('Fail')
```

## 3. Nested If-Else Statements

```python
num = int(input('Enter a number: '))
if num >= 0:
    if num % 2 == 0:
        print('Positive Even')
    else:
        print('Positive Odd')
else:
    print('Negative number')
```

```python
x, y = 5, 10
if x > 0:
    if y > 0:
        print('Both are positive')
    else:
        print('x positive, y non-positive')
else:
    print('x is non-positive')
```

```python
marks = 85
if marks >= 50:
    if marks >= 80:
        print('Grade: A')
    else:
        print('Grade: B')
else:
    print('Fail')
```

## 4. Functions

```python
def greet(name):
    print('Hello', name)

user_name = input('Enter your name: ')
greet(user_name)
```

```python
def add(a, b):
    return a + b
print('Addition:', add(5, 3))
```

```python
def square(x):
    return x * x
print('Square:', square(4))
```

```python
def power(base, exp=2):
    return base ** exp
print('Power with default:', power(3))
print('Power with custom exp:', power(3, 3))
```

```python
def Formula(a, b, exp=2):
    return (a+b) ** exp
print('Formula:', Formula(2, 3))
```

## 5. For Loops

```python
for i in range(5):
    print('Iteration:', i)
```

```python
n = int(input('Enter a number: '))
for i in range(1, n+1):
    print(i)
```

```python
fruits = ('apple', 'banana', 'cherry')
for fruit in fruits:
    print('Fruit:', fruit)
```

```python
for char in 'Python':
    print('Character:', char)
```

```python
for i in range(2, 11, 2):
    print('Even number:', i)
```

## 6. NumPy Examples

```python
import numpy as np
arr = np.array([1, 2, 3, 4])
print('Array:', arr)
print('Mean:', np.mean(arr))
```

```python
arr = np.zeros((2,3))
print('Zeros Array:\n', arr)
```

```python
arr = np.ones((3,3))
print('Ones Array:\n', arr)
```

```python
arr = np.arange(0,10,2)
print('Arange:', arr)
```

```python
arr = np.array([1,2,3])
print('Sum:', np.sum(arr))
```

```python
arr = np.array([[1,2],[3,4]])
print('Matrix Transpose:\n', arr.T)
```

```python
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
print('Elementwise Addition:', arr1 + arr2)
```
