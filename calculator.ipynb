```python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def div(a, b):
    if b == 0:
        return "Error! Division by zero."
    return a / b


x = int(input('Enter first number: '))
op = input('Enter operation you want to perform ==> + , - , * , / : ')
y = int(input('Enter second number: '))

if op == '+':
    result = add(x, y)

elif op == '-':
    result = subtract(x, y)

elif op == '*':
    result = multiply(x, y)

elif op == '/':
    result = div(x, y)

else:
    result = "Invalid operator!"

print("Result:", result)
```

```python
#using deictionary
operations = {
    '+': add,
    '-': subtract,
    '*': multiply,
    '/': div
}

if op in operations:
    result = operations[op](x, y)
else:
    result = "Invalid operator!"

print("Result:", result)
```
