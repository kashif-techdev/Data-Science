```python
# LIST DEMONSTRATION

my_list = [10, 20, 20, 30]

print("Original List:", my_list)

# Mutable: change element
my_list[1] = 99
print("After modification:", my_list)

# Add element
my_list.append(40)
print("After append:", my_list)

# Duplicates allowed
my_list.append(40)
print("Duplicates allowed:", my_list)
```

```python
# TUPLE DEMONSTRATION

my_tuple = (10, 20, 20, 30)

print("Tuple:", my_tuple)

# Access allowed
print("Element at index 1:", my_tuple[1])

# Try to modify (will cause error)
# my_tuple[1] = 99   # ‚ùå TypeError

print("Duplicates allowed in tuple:", my_tuple)
```

```python
# DICTIONARY DEMONSTRATION

my_dict = {
    "name": "Kashif",
    "age": 21,
    "age": 22  # duplicate key (overwritten)
}

print("Dictionary:", my_dict)

# Access value using key
print("Name:", my_dict["name"])

# Modify value
my_dict["age"] = 23
print("After modification:", my_dict)

# Add new key-value
my_dict["city"] = "Islamabad"
print("After adding new key:", my_dict)
```

```python
# SET DEMONSTRATION

my_set = {10, 20, 20, 30, 40}

print("Set:", my_set)

# Add element
my_set.add(50)
print("After adding element:", my_set)

# Remove element
my_set.remove(20)
print("After removal:", my_set)

# Duplicate automatically removed
print("Unique elements only:", my_set)
```
